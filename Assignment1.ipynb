```python
#Weather-Based Daily Recommendation Generator

import tracery
from tracery.modifiers import base_english

rules = {
    "origin": "Today is #weather.capitalize#, #recommendation#",
    "weather": [
        "sunny",
        "rainy",
        "windy",
        "snowy",
        "cloudy"
    ],
    "recommendation": [
        "Perfect time to #activity# and enjoy some #drink#.",
        "Maybe stay inside and #indoor# with a cup of #drink#.",
        "A great day to #outdoor# and don’t forget your #item#.",
        "You might want to #relax# while listening to #music#."
    ],
    "activity": ["go for a walk", "read a book", "take some photos", "call a friend"],
    "indoor": ["watch Netflix", "bake some cookies", "plan your next trip", "organize your space"],
    "outdoor": ["visit the park", "ride your bike", "go for a run", "explore a new neighborhood"],
    "relax": ["light a candle", "do some stretching", "write a diary", "just breathe deeply"],
    "drink": ["latte", "green tea", "hot chocolate", "coffee"],
    "item": ["smartphone", "scarf", "camera", "airpods"],
    "music": ["pop", "jazz", "k-pop", "classical piano"]
}

grammar = tracery.Grammar(rules)
grammar.add_modifiers(base_english)

for i in range(10):
    print(grammar.flatten("#origin#"))
```
