```python
import spacy

# ëª¨ë¸ ë¡œë“œ
nlp = spacy.load("en_core_web_md")

print("Word Similarity Recommender")
print("Enter a word to get similar word suggestions. (Type 'quit' to exit)\n")

while True:
    word = input("ë‹¨ì–´ ì…ë ¥: ").strip()

    if word == "quit":
        print("Program terminated")
        break

    if word not in nlp.vocab:
        print("This word is not in the vocabulary. Please enter another word.\n")
        continue

    token = nlp.vocab[word]

    if not token.has_vector:
        print("This word does not have a word vector, so similar words cannot be found.\n")
        continue

    similarities = []
    for t in nlp.vocab:
        if t.has_vector and t.is_alpha and t.is_lower:
            sim = token.similarity(t)
            similarities.append((t.text, sim))

    similarities = sorted(similarities, key=lambda x: x[1], reverse=True)

    top5 = similarities[1:6]

    print(f"\nğŸ“Œ Top 5 words similar to '{word}':")
    for w, s in top5:
        print(f" - {w:15s}  (similarity: {s:.4f})")
    print("\n")
```
